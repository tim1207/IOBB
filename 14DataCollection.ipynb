```python
import serial
import csv
import time
import pandas as pd
from datetime import datetime, timedelta
from time import sleep
serialPort = serial.Serial(port='COM3', baudrate=57600, timeout=2, parity=serial.PARITY_EVEN, stopbits=1)
serialPort2 = serial.Serial(port='COM5', baudrate=57600, timeout=2, parity=serial.PARITY_EVEN, stopbits=1)
#/dev/cu.name
size = 1024
xg,yg,zg,mx,my,mz,xa,ya,za,xq,yq,zq,direc,mytime =[],[],[],[],[],[],[],[],[],[],[],[],[],[]
xg2,yg2,zg2,mx2,my2,mz2,xa2,ya2,za2,xq2,yq2,zq2,direc2,mytime2 =[],[],[],[],[],[],[],[],[],[],[],[],[],[]
Ttime=str(datetime.now())
print(Ttime[17::])
tt=0
while(tt<=10):
    serialPort.write("csdcsdc\n".encode('utf-8'))
#     sleep(1)
    data = serialPort.readline().decode('utf-8')
    data2 = serialPort2.readline().decode('utf-8')
    print(data2)
    
    data=data.split(',')
    data2=data2.split(',')
    
    if (float(str(datetime.now())[17::])< float(Ttime[17::])):      
        tt = float(Ttime[17::])+60-float(str(datetime.now())[17::] )
        if tt>=10:
            break
    else:
        tt = -float(Ttime[17::])+float(str(datetime.now())[17::] )
        if tt>=10:
            break
    if len(data)==15:
        mx.append(data[0])
        my.append(data[1])
        mz.append(data[2])
        xg.append(data[3])
        yg.append(data[4])
        zg.append(data[5])
        xa.append(data[6])
        ya.append(data[7])
        za.append(data[8])
        xq.append(data[9])
        yq.append(data[10])
        zq.append(data[11])
        direc.append(data[12])
        mytime.append(data[13])
    if len(data2)==15:
        mx2.append(data2[0])
        my2.append(data2[1])
        mz2.append(data2[2])
        xg2.append(data2[3])
        yg2.append(data2[4])
        zg2.append(data2[5])
        xa2.append(data2[6])
        ya2.append(data2[7])
        za2.append(data2[8])
        xq2.append(data2[9])
        yq2.append(data2[10])
        zq2.append(data2[11])
        direc2.append(data2[12])
        mytime2.append(data2[13])
pandasData={"Xg":xg,"Yg":yg,"Zg":zg,"Xm":mx,"Ym":my,"Zm":mz,"Xa":xa,"Ya":ya,"Za":za,"Xq":xq,"Yq":yq,"Zq":zq,"Direction":direc,"Time":mytime}
pandasData2={"Xg":xg2,"Yg":yg2,"Zg":zg2,"Xm":mx2,"Ym":my2,"Zm":mz2,"Xa":xa2,"Ya":ya2,"Za":za2,"Xq":xq2,"Yq":yq2,"Zq":zq2,"Direction":direc2,"Time":mytime2}
df = pd.DataFrame(pandasData)
df2= pd.DataFrame(pandasData2)
print(df)
print(df2)
df.to_csv('蹺蹺板1'+datetime.now().strftime('%Y-%m-%d %H%M%S')+'.csv')
df2.to_csv('蹺蹺板2'+datetime.now().strftime('%Y-%m-%d %H%M%S')+'.csv')
serialPort.close()
serialPort2.close()
```
