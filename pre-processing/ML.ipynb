```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

np.random.seed(202205456)
```

```python
def SVM(data):     
    data = np.array(pd.read_csv(data))
    labelSeries = data[:,16]
    dataSeries = data[:,1:10]
    x_train, x_test, y_train, y_test = train_test_split(dataSeries, labelSeries,test_size=0.2, random_state=2222)


    svm = SVC(kernel="rbf",gamma = 0.5, probability=True)
    svm.fit(x_train,y_train)

    predict = svm.predict(x_test)
    svm_score = svm.score(x_test, y_test) 
    print(classification_report(y_test,predict))
    return svm_score
```

```python
def DT(data):
    data = np.array(pd.read_csv(data))
    labelSeries= data[:,16]
    dataSeries = data[:,1:10]
    
    # 使用訓練資料訓練模型
    x_train, x_test, y_train, y_test = train_test_split(dataSeries, labelSeries,test_size=0.2, random_state=211)

    decisionTreeModel = DecisionTreeClassifier()
    decisionTreeModel.fit(x_train, y_train)
    # 使用訓練資料預測

    predict= decisionTreeModel.predict(x_test)
    
    DT_score = decisionTreeModel.score(x_test,y_test)
    print(classification_report(y_test,predict))
    return DT_score
```

```python
def RF(data):
    data = np.array(pd.read_csv(data))
    labelSeries = data[:,16]
    dataSeries = data[:,1:10]
    x_train, x_test, y_train, y_test = train_test_split(dataSeries, labelSeries,test_size=0.2, random_state=222)
    
    randomForestModel = RandomForestClassifier(n_estimators=100, criterion = 'gini')
    # 使用訓練資料訓練模型
    randomForestModel.fit(x_train, y_train)
    # 使用訓練資料預測
    predict = randomForestModel.predict(x_test)
    RF_score = randomForestModel.score(x_test,y_test)
    print(classification_report(y_test,predict))
    return RF_score
```

```python
# Support Vector Machine支援向量機
print("--------Support Vector Machine ---------")
for i in range(len(data0510)//2):
    print("Hand:",SVM(data0510[i]))
    print("")
    print("Foot:",SVM(data0510[i+3]))
    print("________________________________________________________")
```

```python
print("-------------------------------\nDecision Treeeeeeeeeeeeeeeeeeee\n-------------------------------")
for i in range(len(data0510)//2):
    print("Hand:",DT(data0510[i]))
    print("")
    print("Foot:",DT(data0510[i+3]))
    print("________________________________________________________")
```

```python
print("-------------------------------\nRandomForest\n-------------------------------")
for i in range(len(data0510)//2):
    print("Hand:",RF(data0510[i]))
    print("")
    print("Foot:",RF(data0510[i+3]))
    print("________________________________________________________")
```
