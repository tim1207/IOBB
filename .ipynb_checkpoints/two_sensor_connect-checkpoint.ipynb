```python
import serial
from datetime import datetime
import csv
import pandas as pd
from time import sleep
serialPort = serial.Serial(port='COM3', baudrate=57600, timeout=2, parity=serial.PARITY_EVEN, stopbits=1)
serialPort2 = serial.Serial(port='COM5', baudrate=57600, timeout=2, parity=serial.PARITY_EVEN, stopbits=1)
#/dev/cu.name
size = 1024
xg,yg,zg,mx,my,mz =[],[],[],[],[],[]
xg2,yg2,zg2,mx2,my2,mz2 =[],[],[],[],[],[]
for i in range (0,300):
    print(i)
    #serialPort.write("csdcsdc\n".encode('utf-8'))
    #sleep(1)
    data = serialPort.readline().decode('utf-8')
    data2 = serialPort2.readline().decode('utf-8')
    print(data)
    print('-')
    print(data2)
    data=data.split(',')
    data2=data2.split(',')
    if len(data)==7:
        mx.append(data[0])
        my.append(data[1])
        mz.append(data[2])
        xg.append(data[3])
        yg.append(data[4])
        zg.append(data[5])
    if len(data2)==7:
        mx2.append(data2[0])
        my2.append(data2[1])
        mz2.append(data2[2])
        xg2.append(data2[3])
        yg2.append(data2[4])
        zg2.append(data2[5])

pandasData={"Xg":xg,"Yg":yg,"Zg":zg,"Xm":mx,"Ym":my,"Zm":mz}
pandasData2={"Xg":xg2,"Yg":yg2,"Zg":zg2,"Xm":mx2,"Ym":my2,"Zm":mz2}
df = pd.DataFrame(pandasData)
print(df)
df1 = pd.DataFrame(pandasData2)
print(df1)
serialPort.close()
serialPort2.close()
df.to_csv('Output'+datetime.now().strftime('%Y-%m-%d %H%M%S')+'.csv')
df1.to_csv('Output1'+datetime.now().strftime('%Y-%m-%d %H%M%S')+'.csv')
```
