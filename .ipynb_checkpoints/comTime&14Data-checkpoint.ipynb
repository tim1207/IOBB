```python
import serial
import csv
import time
import pandas as pd
from datetime import datetime, timedelta
from time import sleep
serialPort = serial.Serial(port='COM7', baudrate=115200, timeout=2, parity=serial.PARITY_EVEN, stopbits=1)
serialPort2 = serial.Serial(port='COM4', baudrate=115200, timeout=2, parity=serial.PARITY_EVEN, stopbits=1)


#/dev/cu.name
size = 1024
comTime,xg,yg,zg,mx,my,mz,xa,ya,za,xq,yq,zq,direc,mytime =[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]
comTime2,xg2,yg2,zg2,mx2,my2,mz2,xa2,ya2,za2,xq2,yq2,zq2,direc2,mytime2 =[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]
Ttime=str(datetime.now())
print(Ttime[17::])
tt=0
while(tt<=10):
    serialPort.write(bytes(datetime.now().strftime('%H:%M:%S')+"#", 'big5'))
    serialPort2.write(bytes(datetime.now().strftime('%H:%M:%S')+"#", 'big5'))
    data = serialPort.readline().decode('utf-8')
    data2 = serialPort2.readline().decode('utf-8')
    print(data)
    print(data2)
    data=data.split(',')
    data2=data2.split(',')
    
    if (float(str(datetime.now())[17::])< float(Ttime[17::])):      
        tt = float(Ttime[17::])+60-float(str(datetime.now())[17::] )
        if tt>=10:
            break
    else:
        tt = -float(Ttime[17::])+float(str(datetime.now())[17::] )
        if tt>=10:
            break
    if len(data)==16:
        comTime.append(data[0])
        mx.append(data[1])
        my.append(data[2])
        mz.append(data[3])
        xg.append(data[4])
        yg.append(data[5])
        zg.append(data[6])
        xa.append(data[7])
        ya.append(data[8])
        za.append(data[9])
        xq.append(data[10])
        yq.append(data[11])
        zq.append(data[12])
        direc.append(data[13])
        mytime.append(data[14])
    if len(data2)==16:
        comTime2.append(data[0])
        mx2.append(data[1])
        my2.append(data[2])
        mz2.append(data[3])
        xg2.append(data[4])
        yg2.append(data[5])
        zg2.append(data[6])
        xa2.append(data[7])
        ya2.append(data[8])
        za2.append(data[9])
        xq2.append(data[10])
        yq2.append(data[11])
        zq2.append(data[12])
        direc2.append(data[13])
        mytime2.append(data[14])
        
pandasData={"Xg":xg,"Yg":yg,"Zg":zg,"Xm":mx,"Ym":my,"Zm":mz,"Xa":xa,"Ya":ya,"Za":za,"Xq":xq,"Yq":yq,"Zq":zq,"Direction":direc,"Time":mytime,"COMTime":comTime}
pandasData2={"Xg":xg2,"Yg":yg2,"Zg":zg2,"Xm":mx2,"Ym":my2,"Zm":mz2,"Xa":xa2,"Ya":ya2,"Za":za2,"Xq":xq2,"Yq":yq2,"Zq":zq2,"Direction":direc2,"Time":mytime2,"COMTime":comTime2}
df = pd.DataFrame(pandasData)
df2= pd.DataFrame(pandasData2)
print(df)
print(df2)
df.to_csv('ComTime&14data'+datetime.now().strftime('%Y-%m-%d %H%M%S')+'.csv')
df2.to_csv('ComTime&14data2'+datetime.now().strftime('%Y-%m-%d %H%M%S')+'.csv')
serialPort.close()
serialPort2.close()
```
