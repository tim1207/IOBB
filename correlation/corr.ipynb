```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

data = pd.read_csv(dataCSV[0])
data2 = pd.read_csv(dataCSV[1])

data = data.drop(["Unnamed: 0",'Time'],axis=1)
data2 = data2.drop(["Unnamed: 0",'Time'],axis=1)

corrMatrix = data.corr(method='pearson') 
corrMatrix2 = data2.corr(method='pearson')

corrMatrix = corrMatrix.replace(np.nan,0)
corrMatrix2 = corrMatrix2.replace(np.nan,0)


sn.set(rc = {'figure.figsize':(40,40)})
sn.set(font_scale=2)
sn.heatmap(corrMatrix, annot=True)
plt.show()


sn.set(rc = {'figure.figsize':(40,40)})
sn.set(font_scale=2)
sn.heatmap(corrMatrix2, annot=True)
plt.show()

sn.set(rc = {'figure.figsize':(40,40)})
sn.set(font_scale=2)
sn.heatmap((corrMatrix+corrMatrix2)/2, annot=True)
plt.show()
```
