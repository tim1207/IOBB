```python
import serial
import csv
import pandas as pd
from datetime import datetime
from time import sleep
import time
serialPort = serial.Serial(port='COM7', baudrate=115200, timeout=2, parity=serial.PARITY_EVEN, stopbits=1)
i=1
def write_read():
    serialPort.write(bytes(datetime.now().strftime('%H%M%S')+"#", 'big5'))
    data = serialPort.readline().decode('utf-8')
    return data
while True:
    print(write_read())
serialPort.close()
```
